- create an empty list to be the container for the data you want to retrieve
- insert data into the container one by one according to the empty list name that was created earlier, with a loop if the data is not found it will be written Unknown or 0
- retrieve data on product name, price, rating, and gender

```python
product_names = []
prices = []
rating = []
genders = []

for container in container_product:
    product_names.append(container.find("h3",{"class":"product-tile-product-description"}).text)
    if container.find("span",{"class":"price-limited-ER"}) != None:
        prices.append(container.find("span",{"class":"price-limited-ER"}).text)
    elif container.find("span",{"class":"price-original-ER"}) != None:
        prices.append(container.find("span",{"class":"price-original-ER"}).text)
    else:
        prices.append("0")
    if container.find("span",{"class":"bold ml-xxxs small-rating"}) != None:
        rating.append(container.find("span",{"class":"bold ml-xxxs small-rating"}).text)
    else:
        rating.append("0")
    if container.find("p",{"class":"product-tile-category-item-gender"}) != None:
        genders.append(container.find("p",{"class":"product-tile-category-item-gender"}).text)
    else:
        genders.append("Unknown")
```

check the data length of each list/series (because if the data length is not the same, the series cannot be merged into a dataframe)

```python
len(product_names), len(prices), len(rating), len(genders)
```

create a dataframe with the name df_uniqlo by inserting all the filled lists/series, and display the top 10 data.

```python
df_uniqlo = pd.DataFrame({"Product Name":product_names,
                              "Prices":prices,
                              "Rating":rating,
                              "Gender":genders
})
df_uniqlo.head(10)
```

B. Transform

Remove the Rp in the 'price' column and change the data type of the prices column to and rating to float 

```python
df_uniqlo['Prices'] = df_uniqlo['Prices'].str.replace('Rp|\.', '', regex=True)
```

```python
df_uniqlo['Prices'] = df_uniqlo['Prices'].astype(int)
```

```python
df_uniqlo['Rating'] = df_uniqlo['Rating'].astype(float)
```
